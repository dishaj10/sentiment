```python
import pandas as pd
```

```python
import nltk #nltk stands for natural language tool kit,that helps machine providing some english words to understand the sentiments of the user
from nltk.corpus import stopwords #corpus is like a bag or a group that is consisted with different kinds of english words.
from nltk.stem import PorterStemmer #PortStemmer is a python library that helps the machine to break the machine to break the words into their root forms.
import string
```

```python
df = pd.read_excel("/content/test.xlsx")
```

```python
df
```

```python
nltk.download('stopwords')
nltk.download('punkt') # Download the punkt tokenizer data
stop_words = stopwords.words("english") #will check for stopwords(non relavent words inside a sentence)
ps=PorterStemmer()
```

```python
def preprocessing(text):
  text = text.lower()
  token = nltk.word_tokenize(text)
  new = []
  for i in token:
    if i not in stop_words and i not in string.punctuation:
      Stemmer = ps.stem(i)
      new.append(Stemmer)
  return " ".join(new)
```

```python
nltk.download('punkt_tab', quiet=True)
df["clean"] = df["text"].apply(preprocessing)
```

```python
import nltk
nltk.download('punkt_tab')
df['clean']=df['text'].apply(preprocessing)
df["clean"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vector = TfidfVectorizer()
x = vector.fit_transform(df['clean'])
y = df['sentiment']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
log=LogisticRegression()
log.fit(X_train,y_train)
```

```python
def pred_text(text):
  text=text.lower()
  text=preprocessing(text)
  vec=vector.transform([text])
  predict=log.predict(vec)[0]
  if predict=='positive':
    return "positive"
  elif predict=='Neutral':
    return "Neutral"
  else:
    return "Negative"
```
